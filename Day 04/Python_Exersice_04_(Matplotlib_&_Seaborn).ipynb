```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("--- Intermediate Matplotlib & Seaborn Exercise ---")
print("Complete each task by writing the requested plotting code.")
print("Focus on proper labeling, titles, and drawing insights.")
print("--------------------------------------------------")

# Load the dataset
# We'll use the 'tips' dataset, which is built into Seaborn.
# It contains information about restaurant tips: total bill, tip, gender of bill payer, smoker status, day, time, and size of party.

tips = sns.load_dataset('tips')
print("Tips Dataset Loaded Sucessfully")
```

```python
# --- Task 1: Single Variable Distributions (Histogram, Pie Chart) ---

print("--- Task 1: Single Variable Distributions (Histogram, Pie Chart---\n")
# 1.1 Histogram: Distribution of 'total_bill'
# Create a histogram of the 'total_bill' column.
# Add a title, x-label, and y-label. Use `bins=20`.

sns.histplot(data=tips, x="total_bill", kde=True)
plt.title("Distribution of Total Bills")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

**Insight Question: What is the typical range of total bills? Is the distribution symmetric or skewed?** <br>
= Typical Range of Total Bills = 10 - 25 <br>     The distribution is skewed to the right (positively skewed)

```python
# 1.2 Pie Chart: Proportion of 'smoker' status
# Create a pie chart showing the proportion of smokers vs. non-smokers.
# Ensure the percentages are displayed on the slices.

smoker_counts = tips['smoker'].value_counts()
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Smokers')
plt.show()
```

**Insight Question: Is there a significant difference in the number of smokers vs. non-smokers in the dataset?** <br>
= Yes, smokers are almost half of non-smokers

```python
# --- Task 2: Relationships Between Two Variables (Scatter, Bar, Horizontal Bar) ---

print("--- Task 2: Relationships Between Two Variables (Scatter, Bar, Horizontal Bar)---\n")
# 2.1 Scatter Plot: 'total_bill' vs. 'tip'
# Create a scatter plot with 'total_bill' on the x-axis and 'tip' on the y-axis.
# Use 'smoker' as the `hue` to differentiate points by smoker status.
# Add a title and labels.

sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker')
plt.title('Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

**Insight Question: Is there a relationship between total bill and tip amount? Does smoker status influence this relationship?** <br>
= Higher total bills lead to higher tips. <br>
Not drastically

```python
# 2.2 Bar Plot: Average 'total_bill' by 'day'
# Create a bar plot showing the average 'total_bill' for each 'day' of the week.
# Add error bars (which Seaborn's barplot does by default, showing confidence intervals).
# Add a title and labels.

sns.barplot(data=tips, x='day', y='total_bill', ci='sd')
plt.title('Average Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Average Total Bill')
plt.show()
```

**Insight Question: Which day has the highest average total bill? Are there noticeable differences between weekdays and weekends?** <br>
= Sunday <br>
Yes weekends have more avg.bills than weekdays.


```python
# 2.3 Horizontal Bar Plot: Average 'tip' by 'time'
# Create a horizontal bar plot showing the average 'tip' for 'Lunch' vs. 'Dinner'.
# Add a title and labels.

plt.figure(figsize=(8, 6))
sns.barplot(data=tips, x='tip', y='time', orient='h', ci='sd')
plt.title('Average Tip by Time of Day')
plt.xlabel('Average Tip')
plt.ylabel('Time of Day')
plt.show()
```

**Insight Question: Do customers tip more during lunch or dinner?** <br>
= Customer tip more in dinner.

```python
# --- Task 3: Multiple Plots and Subplots ---

print("--- Task 3: Multiple Plots and Subplots ---")

# 3.1 Create a figure with 2 rows and 2 columns of subplots.
#    - Top-left: Histogram of 'tip' amount.
#    - Top-right: Box plot of 'total_bill' by 'sex'.
#    - Bottom-left: Count plot of 'day'.
#    - Bottom-right: Swarm plot of 'tip' by 'smoker' status.
# Ensure each subplot has an appropriate title and labels. Use `plt.tight_layout()` to prevent overlap.
# Insight Question: What can you learn about tip distribution, total bill distribution across genders, daily customer patterns, and tip differences based on smoker status from these combined plots?

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Top-left: Histogram of 'tip' amount.
sns.histplot(data=tips, x="tip", kde=True, ax=axs[0, 0])
axs[0, 0].set_title("Distribution of Tip")
axs[0, 0].set_xlabel("Tip Amount")
axs[0, 0].set_ylabel("Frequency")

# Top-right: Box plot of 'total_bill' by 'sex'.
sns.boxplot(data=tips, x="sex", y="total_bill", ax=axs[0, 1])
axs[0, 1].set_title("Total Bill by Sex")
axs[0, 1].set_xlabel("Sex")
axs[0, 1].set_ylabel("Total Bill")

# Bottom-left: Count plot of 'day'.
sns.countplot(data=tips, x="day", ax=axs[1, 0])
axs[1, 0].set_title("Count of Customers by Day")
axs[1, 0].set_xlabel("Day of the Week")
axs[1, 0].set_ylabel("Number of Customers")

# Bottom-right: Swarm plot of 'tip' by 'smoker' status.
sns.swarmplot(data=tips, x="smoker", y="tip", ax=axs[1, 1])
axs[1, 1].set_title("Tip Amount by Smoker Status")
axs[1, 1].set_xlabel("Smoker")
axs[1, 1].set_ylabel("Tip Amount")

plt.tight_layout()
plt.show()
```

**Insight Question: What can you learn about tip distribution, total bill distribution across genders, daily customer patterns, and tip differences based on smoker status from these combined plots?** <br>

1. Tips are also right-skewed, with most tips being smaller amounts (2-4).
2. While both genders have a similar median total bill, males seem to have a slightly wider spread and potentially more outliers on higher end of total bills.
3. Saturday is the busiest day, followed by Sunday and Thursday, with Friday being the least busy among the days recorded.
4. Visually, it appears that non-smokers might give a slightly wider range of tips, including some higher tips, though the bulk of tips for both groups falls in a similar range. There's no immediately obvious huge difference in average tip based on smoker status from the swarm plot alone.
