```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#For graphics
%matplotlib inline
```

**1. Line Plot** - plt.plot()

```python
x = np.linspace(0, 10, 100)
y = np.tan(x)

plt.figure(figsize=(8, 4))
plt.plot(x,y,label = 'sin(x)', color='red', linewidth=2)
plt.title('Line Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
```

**2. Bar Graph** - plt.bar()

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 50, 100, 200]

plt.figure(figsize=(8, 4))
plt.bar(categories, values, color='orange')
plt.title('Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

**3. Horizontal Bar Graph** - plt.barh()

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 50, 100, 200]

plt.figure(figsize=(8, 4))
plt.barh(categories, values, color='yellow')
plt.title('Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

**4. Scatter Plot** - plt.scatter()

```python
x = np.random.rand(50)
y = x+np.random.normal(0, 0.1, 50)

plt.figure(figsize = (6,4))
plt.scatter(x, y, color='green')
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**5. Histogram** - plt.hist

```python
data = np.random.randn(1000)
plt.figure(figsize=(6,4))
plt.hist(data,bins=30, color='skyblue',edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

**6. Pie-Chart** - plt.pie()

```python
sizes = [10, 20, 15, 20, 10, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
colors = ['red', 'orange', 'yellow', 'green', 'skyblue','blue', 'violet']

plt.figure(figsize=(6,4))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.axis('equal')
plt.show()
```

**7. Multiple Plots** - plt.plot()

```python
x = np.linspace(0, 2*np.pi, 400)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8,4))
plt.plot(x,y1, label='sin(x)', color = 'brown')
plt.plot(x,y2, label='cos(x)', color = 'purple')
plt.title('Multiple Plots')
plt.xlabel('x')
plt.ylabel('sin(x), cos(x)')
plt.legend()
plt.show()
```

**8. Subplots**

```python
fig, axs = plt.subplots(1,2, figsize=(10,4))
axs[0].plot(x,y1, label='sin(x)', color = 'brown')
axs[0].set_title('sin(x)')
axs[1].plot(x,y2, label='cos(x)', color = 'purple')
axs[1].set_title('cos(x)')
plt.suptitle('Subplots')
plt.show()
```

**9. Boxplot** - plt.boxplot()

```python
np.random.seed(5)
data = np.random.randint(10, 100, 100)

plt.figure(figsize=(8,4))
plt.boxplot(data, vert=True)
plt.title('Boxplot')
plt.show()
```

**10. Area Plot** - plt.stackplot()

```python
days = np.arange(1,8)
milk = np.random.randint(1,10, 7)
water = np.random.randint(1,10, 7)
coffee = np.random.randint(1,10, 7)

plt.figure(figsize=(8,4))
plt.stackplot(days, milk, water, coffee, labels=['Milk', 'Water', 'Coffee'], colors=['blue','indigo','violet'])
plt.xlabel('Days')
plt.ylabel('Units')
plt.legend()
plt.title('Area Plot')
plt.show()
```

**11. Error Bar Plot** - plt.errorbar()

```python
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)
error = 0.1 + 0.2*x

plt.figure(figsize=(6,4))
plt.errorbar(x,y,yerr=error, fmt='-o', ecolor='red', capsize=5)
plt.xlabel('X')
plt.ylabel('Y = exp(-x)')
plt.title('Error Bar Plot')
plt.show()
```

**12. Categorical Plots**

```python
# Create Synthetic Dataframe with categorical data
category_data = pd.DataFrame({
    'Department' : ['HR','IT','HR','Sales','Sales','IT','IT','HR','Sales', 'IT'],
    'Satisfaction' : [3, 4, 2, 5, 4, 3, 4, 2, 5, 4]
})

# Count plot = Bar chart
dept_count = category_data['Department'].value_counts()
plt.figure(figsize=(8,4))
plt.bar(dept_count.index, dept_count.values, color = 'lightgreen')
plt.title('Department Count')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

```python
mean_satisfaction = category_data.groupby('Department')['Satisfaction'].mean()
plt.figure(figsize=(6, 4))
plt.bar(mean_satisfaction.index, mean_satisfaction.values, color='green')
plt.title('Average Satisfaction by Department')
plt.xlabel('Department')
plt.ylabel('Average Satisfaction')
plt.show()
```
